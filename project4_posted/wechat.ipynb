```python
import pandas as pd
import numpy as np

pd.set_option('max_rows', 50)
```

```python
raw = pd.read_csv('wechat_raw.csv')
raw.head()
```

```python
red = raw[raw['Type']=='Red Packet']
red['Time'] = pd.to_datetime(red['Time'])
red = red.set_index('Time')
```

```python
receive = red[red['Amount']>0].sort_index()
handout = red[red['Amount']<=0].sort_index()
```

```python
receiveD = receive.resample('D', how='sum')
receiveD['Amount'] = receiveD['Amount'].fillna(0)
receiveD = receiveD.cumsum()
```

```python
receiveD['DayofYear'] = receiveD.index.dayofyear
```

```python
receiveD['Year'] = receiveD.index.year
```

```python
receiveD['TotalDays'] = receiveD['Year'].apply(lambda x: 366 if x==2016 else 365)
```

```python
receiveD['NumberofYear'] = receiveD['Year'].apply(lambda x:x-2015)
```

```python
receiveD['Degree'] = receiveD['DayofYear'] / receiveD['TotalDays'] * 360
```

```python
receiveD['sin'] = np.sin(receiveD['Degree'] * np.pi / 180)
```

```python
receiveD['cos'] = np.cos(receiveD['Degree'] * np.pi / 180)
```

```python
r = 10
```

```python
receiveD['x'] = receiveD['sin'] * r
receiveD['y'] = receiveD['cos'] * r
```

```python
receiveD['z'] = receiveD['Amount'] / 50
```

```python
receiveD
```

```python
for index, row in receiveD.iterrows():
    print('-p {0:.2f} {1:.2f} {2:.2f}'.format(row['x'], row['y'], row['z']))
```
